```python
#lecture du dataset sous forme de série temporelle en prenant comme indice du dataset la date (jour-mois-année)
data = pd.read_csv("C:/Users/dell/Downloads/BTC-USD.csv",index_col='Date',parse_dates=True)
```

```python
data.head()
```

```python
#dimension du dataset
data.shape
```

```python
#utiliser matplotlib pour afficher la courbe du bitcoin avec comme indice une date 
data['Close'].plot(figsize=(10,6))
```

```python
data.index
```

```python
# la fonction loc permet de localiser (se focaliser sur ) une partie du dataset pour une meilleure précision sur les données.
#dans ce cas on étudie uniqueent le mos de septembre 2019
data.loc['2019-9','Close'].plot()
```

```python
#afficher la variation de la valeur du bitcoin entre 2019 et 2021
data.loc['2019':'2021','Close'].plot()
```

```python
# utiliser la fonction resample pour montrer les variation sur des intervalles d'un mois
data.loc['2019':'2021','Close'].resample('m').plot()
```

```python
#lisser le resultat par ula fonction mean (moyenne sur un mois). dans ce cas un mois est représenté par une seule valeur égale à la moyenne
data.loc['2019':'2021','Close'].resample('m').mean().plot()
```

```python
#utiliser matplotlib pour construire des figures plus complète
#choisir la taille des figures
plt.figure(figsize=(10,6))
#insérer plusieurs courbes dans la meme figure
data.loc['2019':'2020','Close'].plot()
data.loc['2019':'2020','Close'].resample('m').mean().plot(label='Moyenne par mois',ls=':',alpha=0.8)
data.loc['2019':'2020','Close'].resample('w').mean().plot(label='Moyenne par semaine',ls='--',alpha=0.8)
#donner un nom àa la figure
plt.title('Cours du bitCoin')
#axes
#plt.xlabel('Date')
plt.ylabel('valeur du bitcoin')
#ajouter une légente
plt.legend()
#enregistrer la figure sur disque
plt.savefig("c:/sim/fig1.jpg")
plt.show()
```

```python
# étudier la variation du cours u bitcoin à l'aide de la fonction std (écart type)
data.loc['2019':'2020','Close'].resample('m').std().plot(label='écart type par mois',c='red',ls='--',alpha=0.8)
```

```python
#construire un nouveay dataframe appelé New et contenant la moyenne sur un mois, l'écart type, le minimum et le maximum
New=data['Close'].resample('m').agg(['mean','std','min','max'])
New.head()
```

```python
#représenter le cours du bitcoin en considérant les valeurs min et max durant la journée
plt.figure(figsize=(10,6))
New['mean'].plot(label='moyenne par semaine')
plt.fill_between(New.index,New['min'],New['max'],alpha=0.3,label='min-max par semaine')
```
